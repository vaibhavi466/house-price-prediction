# Exploratory Data Analysis - Bangalore House Prices

This notebook explores the cleaned house prices dataset. It analyzes the features, distributions, correlations, and outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## Load Cleaned Dataset

```python
df = pd.read_csv('../data/processed/cleaned_house_data.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## Basic Summary Statistics

```python
df.describe()
```

## Distribution of Prices

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='royalblue')
plt.title('Distribution of House Prices in Bengaluru', fontsize=14)
plt.xlabel('Price (Lakhs INR)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

## Distribution of BHK (Bedrooms)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='bhk', data=df, color='teal')
plt.title('Distribution of Bedrooms (BHK)', fontsize=14)
plt.xlabel('BHK', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

## Total Square Feet vs Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_sqft', y='price', hue='bhk', data=df, palette='viridis', alpha=0.7)
plt.title('Total Square Feet vs Price', fontsize=14)
plt.xlabel('Total Square Feet', fontsize=12)
plt.ylabel('Price (Lakhs INR)', fontsize=12)
plt.show()
```

## Bathrooms vs BHK Relationship

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bhk', y='bath', data=df, color='coral')
plt.title('Bathrooms by BHK Tier', fontsize=14)
plt.xlabel('BHK', fontsize=12)
plt.ylabel('Bathrooms', fontsize=12)
plt.show()
```

## Correlations

```python
plt.figure(figsize=(8, 6))
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Features', fontsize=14)
plt.show()
```
